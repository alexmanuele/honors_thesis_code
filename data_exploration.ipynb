```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_table('rel_abu.from_biom.txt', sep='\t',encoding='utf-8')
data = data.T
```

```python
data.columns = data.iloc[0]
data = data.drop(labels='ID')
data.head()
```

```python
data.to_csv('relative_abundance.csv')
```

```python
data = data.sort_index()
rows = data.index
sub = []
supra = []
for i in rows:
    if "SubG" in i:
        sub.append(i)
    if "SupG" in i:
        supra.append(i)
sub_data = data.loc[sub]
supra_data = data.loc[supra]

sub_data.to_csv('all_subg_samples.csv')
supra_data.to_csv('all_supg_samples.csv')

day_0 = []
day_35 = []
for i in rows:
    if "-0" in i:
        day_0.append(i)
    if "-35" in i:
        day_35.append(i)

init_data = data.loc[day_0]
final_data = data.loc[day_35]

init_data.to_csv('all_init_data.csv')
final_data.to_csv('all_final_data.csv')

```

```python
sub = pd.read_csv('all_subg_samples.csv', index_col=0)
sup = pd.read_csv('all_supg_samples.csv', index_col=0)

sub_rows = sub.index
sub_grouped = []
for i in sub_rows:
    if "-14" in i or "-21" in i or "-28" in i:
        sub_grouped.append(i)
        
sup_rows = sup.index
sup_grouped = []
for i in sup_rows:
    if "-14" in i or "-21" in i or "-28" in i:
        sup_grouped.append(i)
        
sub_selected = sub.loc[sub_grouped]
sup_selected = sup.loc[sup_grouped]

sub_selected.to_csv('sub_selected.csv')
sup_selected.to_csv('sup_selected.csv')
```
